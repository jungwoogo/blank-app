```python
from IPython.display import display, Markdown
import ipywidgets as widgets

def recommend_crrt_fluid(serum_k, serum_p):
    recommendation = {}
    recommendation['post'] = 'ğŸ’§ Hemosol BO (í•­ìƒ ê³ ì •)'

    if serum_k <= 4.5:
        if serum_p > 3.5:
            recommendation['pre'] = 'ğŸŸ¦ Multibic 4K'
            recommendation['dialysate'] = 'ğŸŸ¦ Multibic 4K'
        else:
            recommendation['pre'] = 'ğŸŸ¨ Phoxilium'
            recommendation['dialysate'] = 'ğŸŸ¨ Phoxilium'
    elif 4.6 <= serum_k <= 5.0:
        recommendation['pre'] = 'ğŸ’§ Hemosol BO'
        recommendation['dialysate'] = 'ğŸŸ¦ Multibic 4K' if serum_p > 3.5 else 'ğŸŸ¨ Phoxilium'
    elif serum_k >= 5.1:
        recommendation['pre'] = 'ğŸ’§ Hemosol BO'
        recommendation['dialysate'] = 'ğŸ’§ Hemosol BO'
    else:
        recommendation['pre'] = recommendation['dialysate'] = recommendation['post'] = 'âŒ Invalid Input'

    return recommendation

# ì œëª© ë° ì•ˆë‚´ ë¬¸êµ¬
display(Markdown("""
# ğŸ¥ CRRT Fluid Recommendation Tool

**ì¤‘í™˜ìì‹¤ ì „ìš© ìë™ ì¶”ì²œ í”„ë¡œê·¸ë¨ì…ë‹ˆë‹¤.**
í™˜ìì˜ Serum Kâº (ì¹¼ë¥¨) ë° P (ì¸ì‚°) ìˆ˜ì¹˜ë¥¼ ì…ë ¥í•˜ë©´
ì ì ˆí•œ CRRT fluid bagì„ ìë™ìœ¼ë¡œ ì¶”ì²œí•´ë“œë¦½ë‹ˆë‹¤.

ğŸ“Œ ê¸°ì¤€ì€ ë³‘ì› ë‚´ë¶€ Protocolì— ë”°ë¼ ì„¤ì •ë˜ì–´ ìˆìŠµë‹ˆë‹¤.
"""))

# ì…ë ¥ ìœ„ì ¯
k_input = widgets.BoundedFloatText(
    value=4.0,
    min=0,
    max=10,
    step=0.1,
    description='ğŸ”¹ Kâº (mEq/L):',
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='10%')
)

p_input = widgets.BoundedFloatText(
    value=3.5,
    min=0,
    max=10,
    step=0.1,
    description='ğŸ”¸ P (mg/dL):',
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='10%')
)

button = widgets.Button(
    description="ğŸ’¡ í™•ì¸",
    button_style='success',
    tooltip='ì…ë ¥í•œ ìˆ˜ì¹˜ ê¸°ì¤€ìœ¼ë¡œ ì¶”ì²œ fluidë¥¼ ê³„ì‚°í•©ë‹ˆë‹¤',
    layout=widgets.Layout(width='30%')
)

output = widgets.Output()

# ë²„íŠ¼ í´ë¦­ ì‹œ ë™ì‘
def on_button_click(b):
    with output:
        output.clear_output()
        k = k_input.value
        p = p_input.value
        result = recommend_crrt_fluid(k, p)

        display(Markdown(f"""
### ğŸ§¾ ì¶”ì²œ ê²°ê³¼

- **Pre-replacement fluid**: {result['pre']}
- **Dialysate fluid**: {result['dialysate']}
- **Post-replacement fluid**: {result['post']}

ğŸ§  *ì£¼ì˜: ìµœì¢… ì˜í•™ì  íŒë‹¨ì€ ì˜ë£Œì§„ì˜ íŒë‹¨ì— ë”°ë¦…ë‹ˆë‹¤.*
"""))

button.on_click(on_button_click)

# êµ¬ì„± í‘œì‹œ
display(widgets.VBox([
    k_input,
    p_input,
    button,
    output
]))
```
