```python
from IPython.display import display, Markdown
import ipywidgets as widgets

def recommend_crrt_fluid(serum_k, serum_p):
    recommendation = {}
    recommendation['post'] = 'üíß Hemosol BO (Ìï≠ÏÉÅ Í≥†Ï†ï)'

    if serum_k <= 4.5:
        if serum_p > 3.5:
            recommendation['pre'] = 'üü¶ Multibic 4K'
            recommendation['dialysate'] = 'üü¶ Multibic 4K'
        else:
            recommendation['pre'] = 'üü® Phoxilium'
            recommendation['dialysate'] = 'üü® Phoxilium'
    elif 4.6 <= serum_k <= 5.0:
        recommendation['pre'] = 'üíß Hemosol BO'
        recommendation['dialysate'] = 'üü¶ Multibic 4K' if serum_p > 3.5 else 'üü® Phoxilium'
    elif serum_k >= 5.1:
        recommendation['pre'] = 'üíß Hemosol BO'
        recommendation['dialysate'] = 'üíß Hemosol BO'
    else:
        recommendation['pre'] = recommendation['dialysate'] = recommendation['post'] = '‚ùå Invalid Input'

    return recommendation

# Ï†úÎ™© Î∞è ÏïàÎÇ¥ Î¨∏Íµ¨
display(Markdown("""
# üè• CRRT Fluid Recommendation Tool

**Ï§ëÌôòÏûêÏã§ Ï†ÑÏö© ÏûêÎèô Ï∂îÏ≤ú ÌîÑÎ°úÍ∑∏Îû®ÏûÖÎãàÎã§.**
ÌôòÏûêÏùò Serum K‚Å∫ (ÏπºÎ•®) Î∞è P (Ïù∏ÏÇ∞) ÏàòÏπòÎ•º ÏûÖÎ†•ÌïòÎ©¥
Ï†ÅÏ†àÌïú CRRT fluid bagÏùÑ ÏûêÎèôÏúºÎ°ú Ï∂îÏ≤úÌï¥ÎìúÎ¶ΩÎãàÎã§.

üìå Í∏∞Ï§ÄÏùÄ Î≥ëÏõê ÎÇ¥Î∂Ä ProtocolÏóê Îî∞Îùº ÏÑ§Ï†ïÎêòÏñ¥ ÏûàÏäµÎãàÎã§.
"""))

# ÏûÖÎ†• ÏúÑÏ†Ø
k_input = widgets.BoundedFloatText(
    value=4.0,
    min=0,
    max=10,
    step=0.1,
    description='üîπ K‚Å∫ (mEq/L):',
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='10%')
)

p_input = widgets.BoundedFloatText(
    value=3.5,
    min=0,
    max=10,
    step=0.1,
    description='üî∏ P (mg/dL):',
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='10%')
)

button = widgets.Button(
    description="üí° ÌôïÏù∏",
    button_style='success',
    tooltip='ÏûÖÎ†•Ìïú ÏàòÏπò Í∏∞Ï§ÄÏúºÎ°ú Ï∂îÏ≤ú fluidÎ•º Í≥ÑÏÇ∞Ìï©ÎãàÎã§',
    layout=widgets.Layout(width='30%')
)

output = widgets.Output()

# Î≤ÑÌäº ÌÅ¥Î¶≠ Ïãú ÎèôÏûë
def on_button_click(b):
    with output:
        output.clear_output()
        k = k_input.value
        p = p_input.value
        result = recommend_crrt_fluid(k, p)

        display(Markdown(f"""
### üßæ Ï∂îÏ≤ú Í≤∞Í≥º

- **Pre-replacement fluid**: {result['pre']}
- **Dialysate fluid**: {result['dialysate']}
- **Post-replacement fluid**: {result['post']}

üß† *Ï£ºÏùò: ÏµúÏ¢Ö ÏùòÌïôÏ†Å ÌåêÎã®ÏùÄ ÏùòÎ£åÏßÑÏùò ÌåêÎã®Ïóê Îî∞Î¶ÖÎãàÎã§.*
"""))

button.on_click(on_button_click)

# Íµ¨ÏÑ± ÌëúÏãú
display(widgets.VBox([
    k_input,
    p_input,
    button,
    output
]))
```
